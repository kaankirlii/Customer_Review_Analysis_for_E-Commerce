```python
import pandas as pd
import numpy as np
import random
```

```python
# Sample data generation
names = ['John', 'Jane', 'Alice', 'Bob', 'Charlie', 'David', 'Emily', 'Frank', 'Grace', 'Hannah', 'Ian', 'Jack', 'Karen', 'Leo', 'Mia', 'Nina', 'Oscar', 'Paul', 'Quincy', 'Rachel', 'Steve', 'Tina', 'Uma', 'Vince', 'Wendy', 'Xander', 'Yara', 'Zane']
surnames = ['Smith', 'Johnson', 'Williams', 'Brown', 'Jones', 'Garcia', 'Miller', 'Davis', 'Rodriguez', 'Martinez', 'Hernandez', 'Lopez', 'Gonzalez', 'Wilson', 'Anderson', 'Thomas', 'Taylor', 'Moore', 'Jackson', 'Martin', 'Lee', 'Perez', 'Thompson', 'White', 'Harris', 'Sanchez', 'Clark', 'Ramirez']
product_names = ['Laptop', 'Smartphone', 'Headphones', 'Smartwatch', 'Tablet', 'Camera', 'Printer', 'Monitor', 'Keyboard', 'Mouse']
comments = [
    "Excellent product, very satisfied!",
    "Pretty good, but could be better.",
    "Not satisfied, the product broke after a week.",
    "Amazing quality and fast delivery!",
    "It's okay, does the job.",
    "Terrible product, won't buy again.",
    "Highly recommend this product.",
    "It's decent for the price.",
    "Best purchase ever!",
    "Wouldn't recommend, not worth the money.",
    "Exceeded my expectations!",
    "Very disappointed with the performance.",
    "Average product, nothing special.",
    "Loving it so far!",
    "Not what I expected.",
    "Five stars, hands down.",
    "Very poor quality.",
    "Worth every penny.",
    "Subpar performance.",
    "Fantastic, will buy again."
]
```

```python
# Generate dataset
data = []
for _ in range(250):
    name = random.choice(names)
    surname = random.choice(surnames)
    product_name = random.choice(product_names)
    rating = random.randint(1, 5)
    comment = random.choice(comments)
    data.append([name, surname, product_name, rating, comment])
```

```python
# Create DataFrame
df = pd.DataFrame(data, columns=["Name", "Surname", "Product name", "Rating", "Comment"])
```

```python
# Save to CSV
df.to_csv('customer_reviews.csv', index=False)
```
